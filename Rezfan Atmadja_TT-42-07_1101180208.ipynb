```
import numpy as np
import matplotlib.pyplot as plt
import math

Bo = 4
rs = np.arange(0,0.004,0.0000001)

Ps = (math.pi**2)*(rs**2)*Bo
Psdbm = 15 * np.log(Ps)

plt.plot(rs, Psdbm, '-')
plt.show()
```
